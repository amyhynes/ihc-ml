```python
import matplotlib.pyplot as plt
import skimage
```

```python
from skimage import data, exposure, io
```

```python
img = io.imread("data/20.tif");
```

```python
for channel in range(img.shape[2]):  # equalizing each channel
    img_equalized[:, :, channel] = exposure.equalize_hist(img[:, :, channel])
```

```python
plt.figure(figsize=(7, 3))

plt.subplot(121)
plt.imshow(img, interpolation='nearest')
plt.axis('off')
plt.subplot(122)
plt.imshow(img_equalized, interpolation='nearest')
plt.axis('off')
plt.tight_layout()
plt.show()
```
